# Центральная предельная теорема своими руками

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import math
import scipy.stats as sts
warnings.filterwarnings('ignore')
```

## Распределение хи-квадрат (chi-squared distribution)

Подробное можно прочитать [здесь](https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D1%85%D0%B8-%D0%BA%D0%B2%D0%B0%D0%B4%D1%80%D0%B0%D1%82).    
Параметры: *k* > 0 - число степенй свободы  
Математическое ожидание: *k*  
Дисперсия : *2k*

Сгенерируем выборку объёма 1000 с распределением хи-квадрат со степенем свободы k = 10.

```python
from scipy.stats import chi2
k = 10 
mean, var, skew, kurt = chi2.stats(k, moments='mvsk')
sample = chi2.rvs(k, size=1000)
```

```python
E = k #математическое ожидание
D = 2*k #дисперсия
print("Математическое ожидание:", E, "\nДисперсия:", D)
```

Далее построим гистограмма выборки и теоретическую плотность распределения:

```python
x = np.linspace(-10,40,1000) #определяем массив точек для построения плотности распределения
plt.hist(sample,  normed=True, label='Sample', color='indigo') #гистограмма
pdf = chi2.pdf(x, k) #вычисляем значения плотности распределения от выборки
plt.ylabel('fraction of samples') #подписываем ось У
plt.xlabel('$x$') #подписываем ось Х
plt.plot(x, pdf, label='theoretical PDF', color='darkgreen', linewidth=2.5) #строим гистрограмму выборки и теоретическую функцию распределения
plt.legend(loc='best')
plt.show()
```

В этом блоке сформируем выборки из 1000 случайных значений, где каждое значение будет определено 5, 10, 30, 50 количество раз. Результаты представим в виде графиков-гистограмм. Поверх каждого графика-гистограммы нарисуем график плотности соответствующего нормального распределения.

### Объем выборки: 5

```python
n = 5
chi2 = sts.chi2(k)

sample_mean_arr = [] #массив выборочных средних
for i in range(1000): #цикл генерации случайных выборок количества n из функции
    sample = chi2.rvs(n) ##генерация выборки
    sample_mean = sum(sample)/n #определение выборочного среднего (в.с.)
    sample_mean_arr.append(sample_mean)#добавляем в массив новое значение в.с.
sample_mean_arr = np.array(sample_mean_arr)

plt.hist(sample_mean_arr, density = True, label="Sample", color="indigo") #строим гистограмму массива средних при n=5
norm = sts.norm(loc=E, scale=np.sqrt(D/n)) #определяем нормальное распределение с расчетными характеристиками хи-квадрат
x = np.linspace(-1, 30, 1000)
plt.plot(x, norm.pdf(x), label="PDF", color="darkgreen", linewidth=2.5) #строим плотность описывающую этот массив нормального распределения
plt.legend(loc='best')
plt.ylabel('fraction of samples') #подписываем ось У
plt.xlabel('$x$') #подписываем ось Х
plt.show()
```
